```python
# Dependencies
import requests
import json
import pandas as pd
from ratelimit import limits, sleep_and_retry
import time

# Retrieve Google API key from config.py
from config import gkey
```

### Import the data from the Wine Reviews csv file

```python
wine_reviews_data = "winemag-data-130k-v2.csv"
wine_reviews_df = pd.read_csv(wine_reviews_data)
wine_reviews_df.head()
print(len(wine_reviews_df))
```

### Create a new dataframe with id and winery name

```python
winery_df = wine_reviews_df[['winery']].copy()
winery_df = winery_df.rename(columns={'winery': 'winery_name'})
winery_df.head()
```

### Get geocoordinates (latitude/longitude) of wineries and store them in a new df

```python
# Create a new empty dataframe that will contain the latitude, longitude, and address
geo_df = pd.DataFrame(columns=['latitude','longitude'])
wineries = winery_df['winery_name'].unique()

# Set limits for calling API
minute = 100
@sleep_and_retry
@limits(calls=4500, period=minute)

# Adding latitude adn longitude to new_geo_df
def call_geo_api(x):
    url = ('https://maps.googleapis.com/maps/api/geocode/json?address={0}&key={1}').format(x, gkey)
    try:
        response = requests.get(url).json()
        json.dump(response)
        # lat = response["results"][0]["geometry"]["location"]["lat"]
        # lng =  response["results"][0]["geometry"]["location"]["lng"]
        # geo_df.at[x,'latitude'] = lat
        # geo_df.at[x,'longitude'] = lng
        return
    except:
        print("Winery not found. Skipping...")

for winery in wineries:
    call_geo_api(winery)

print("finished")
```
